```python
import pandas as pd
import numpy as np
from pathlib import Path

csv_path = Path("..") / "Dataset" / "Superstore_Cleaned.csv"
df = pd.read_csv(csv_path, encoding="latin1")
```

```python
customers = df[
    ["Customer ID","Customer Name","Segment"]
].drop_duplicates()

customers.shape
```

Product Table

```python
products = df[
    [
        "Product ID",
        "Product Name",
        "Category",
        "Sub-Category"
    ]
].drop_duplicates()

products.shape
```

Geography Table

```python
geography = df[
    [
        "Postal Code",
        "City",
        "State",
        "Region",
        "Country"
    ]
].drop_duplicates()

geography.shape
```

Orders Table

```python
orders = df[
    [
        "Order ID",
        "Order Date",
        "Ship Date",
        "Ship Mode",
        "Customer ID",
        "Product ID",
        "Postal Code",
        "Sales",
        "Quantity",
        "Discount",
        "Profit"
    ]
]
```

Save All Tables

```python
customers.to_csv(
    "../Dataset/customers.csv",
    index=False
)

products.to_csv(
    "../Dataset/products.csv",
    index=False
)

geography.to_csv(
    "../Dataset/geography.csv",
    index=False
)

orders.to_csv(
    "../Dataset/orders.csv",
    index=False
)
```
